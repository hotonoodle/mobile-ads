```python
import pandas as pd
import random
import datetime
import csv
```

How to generate a random date between two dates in Python
https://www.adamsmith.haus/python/answers/how-to-generate-a-random-date-between-two-dates-in-python

```python
#generating 10000 fake dates


fake_dates = []

for x in range(100000):
    start_date = datetime.date(2020, 1, 1)
    end_date = datetime.date(2022, 2, 1)

    time_between_dates = end_date - start_date
    days_between_dates = time_between_dates.days

    random_number_of_days = random.randrange(days_between_dates)
    random_date = start_date + datetime.timedelta(days=random_number_of_days)
    fake_dates.append(str(random_date))
    
print(fake_dates)
```

```python
#read mobile-apps csv and generate random row from the list

with open('mobile-apps.csv', encoding='utf-8') as f:
    reader = csv.reader(f)
    app_list = list(reader)
    
```

```python
fake_apps = []

for x in range(100000):
    fake_apps.append(random.choice(app_list))
```

```python
fake_decisions = []

for x in range(100000):
    fake_decisions.append(random.choice(policy_list))
```
